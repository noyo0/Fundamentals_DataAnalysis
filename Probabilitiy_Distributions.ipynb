#### Uniform distribution

$$ f(x) = \begin{cases}
\frac{1}{b - a} & \text{for } a \leq x \leq b \\
0 & \text{for } x < a \text{ or } x > b
\end{cases} $$

>*f(x)* is the probability density function, which represents the probability of a random variable falling within a specific range [a, b].
<br>*a* is the lower limit of the range.
<br>*b* is the upper limit of the range.

or<br>
f(x) = 1 / (b - a)  for a ≤ x ≤ b
<br>f(x) = 0  otherwise


A uniform distribution, also known as a rectangular distribution, is a probability distribution where all outcomes have an equal likelihood of occurring. In a continuous uniform distribution, the probability density is constant over a specified range.

A real-life example of a uniform distribution is the random selection of a number between 1 and 6 when rolling a fair six-sided die. Each number (1, 2, 3, 4, 5, and 6) has an equal probability of 1/6.

```python
# Dice roll modelling
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate random data for simulating dice rolls (1 to 6)
    # random.uniform(low=0.0, high=1.0, size=None) - (includes low, but excludes high).
    # #Draw samples from a uniform distribution.
    # Any value within the given interval is equally likely to be drawn by uniform.

data = np.random.uniform(1, 7, 1000) 

# Create a DataFrame for the data using pandas
df = pd.DataFrame(data, columns=['Dice'])

# Plot a histogram of the generated data
plt.hist(df['Dice'], bins=6, density=True, alpha=0.7, color='b', edgecolor='black')
plt.xlabel('Dice')
plt.ylabel('Probability Density')
plt.title('Uniform Distribution')
plt.show()
```

In this code, we use np.random.randint(1, 7, 1000) to generate 1000 random data points representing dice rolls between 1 and 6. Then, we create a pandas DataFrame and plot a histogram to visualize the distribution of the dice rolls.

### Normal or Gaussian distribution

$$ f(x) = \frac{1}{{\sigma \sqrt{2\pi}}} \, e^{-\frac{{(x - \mu)^2}}{{2\sigma^2}}}$$

>f(x): This is the probability density function, which represents the probability of a random variable 'x' having a particular value.
<br>mu(&mu;): The mean (average) of the distribution.
<br>sigma(&sigma;): The standard deviation, which measures the spread or variability of the distribution.
<br>e: The mathematical constant approximately equal to 2.71828.
<br>pi(&pi;): The mathematical constant pi, approximately equal to 3.14159.

A normal (Gaussian) distribution is a probability distribution that forms a symmetric bell-shaped curve. In this distribution, data tends to cluster around the mean, and its shape is characterized by a mean and standard deviation.

Real-life example: Heights of people in a population often follow a normal distribution. Most people are near the average height, and fewer people are much taller or shorter.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate heights of people in a population with a normal distribution
mean_height = 170  # Mean height in centimeters
std_deviation = 10  # Standard deviation

# Generate random data for heights
heights = np.random.normal(mean_height, std_deviation, 1000)

# Create a DataFrame for the data using pandas
df = pd.DataFrame(heights, columns=['Height'])

# Create a figure
plt.figure(figsize=(8, 6))
plt.title('Height Distribution (Histogram and Normal Curve)')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')

# Plot the histogram
plt.hist(df['Height'], bins=20, density=True, alpha=0.7, color='b', edgecolor='black', label='Histogram')

# Demonstrate Normal distribution curve, 
    # Calculate the PDF values for the normal distribution
x = np.linspace(mean_height - 3 * std_deviation, mean_height + 3 * std_deviation, 100)
pdf = norm.pdf(x, mean_height, std_deviation)

# Plot the normal distribution curve
plt.plot(x, pdf, 'r-', label='Normal Curve')

plt.legend()
plt.grid(True)
plt.show()

```

### Binominal Distribution

A binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials (experiments) where each trial has two possible outcomes: success or failure. The distribution is characterized by two parameters:

n: The number of trials, which is a fixed, predetermined value.
p: The probability of success on an individual trial, which remains constant across all trials.
The probability mass function (PMF) of the binomial distribution is given by:

<br>P(X = k) = (n choose k) * p^k * (1 - p)^(n - k)
<br>

>**P(X = k)**: This represents the probability of getting exactly 'k' successes in 'n' trials. <br>
**(n choose k)** : This is the binomial coefficient, often denoted as C(n, k), and it calculates the number of ways to choose 'k' successes from 'n' trials.<br>
**p^k**: This term represents the probability of 'k' successes.<br>
**(1 - p)^(n - k)**: This term represents the probability of 'n - k' failures.<br>

<br>The binomial distribution is widely used to model scenarios involving a fixed number of trials with two possible outcomes and a constant probability of success.

**Example:**
Flipping a coin with 50% probability (p = 0.5), 5 times (n = 5).<br> The binomial distribution can be used to model the probability of getting a specific number of heads (successes) in those 5 coin flips.

<br>
P(X = 3) = (5 choose 3) * (0.5)^3 * (0.5)^(5 - 3)
         = (10) * (0.125) * (0.125)
         = *0.3125*
<br> the probability of getting 3 heads in 5 coin flips is 0.3125




```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate a binomial distribution
n_trials = 10 #10 tries
p_success = 0.5 #50% chance
num_simulations = 10000  # Number of simulations

# Simulate the binomial distribution
results = np.random.binomial(n_trials, p_success, num_simulations)

# Create a histogram
plt.figure(figsize=(8, 6))
# the +2 is added to make sure all possible outcomes are covered
 # np.arange(0, n_trials + 2) creates an array that goes from 0 to 11 
plt.hist(results, bins=np.arange(0, n_trials + 2))

plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xticks(range(n_trials + 1)) # cover all 10 outcomes again
plt.grid(True)
plt.show()
```

The resulting histogram resembles a normal distribution histogram but only because the probability (p) was set to 50%.

>Central Limit Theorem: The Central Limit Theorem states that the sampling distribution of the sample mean from a population with any shape of distribution approaches a normal distribution as the sample size increases. In this case, the number of heads in 10 coin tosses represents a sample mean. As the number of simulations (sample size) increases in the code, the histogram will tend to approximate a normal distribution.
