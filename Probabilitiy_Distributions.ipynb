## Types of different data distributions (Histogram)

Histograms are graphical representations of data distributions, and the choice of the appropriate probability distribution to model them depends on the characteristics of the data. Here are some common types of histograms and the probability distributions that can be used to model them:

1. **Normal Distribution (Gaussian)**:
   - **Description**: A symmetric, bell-shaped distribution with a single peak at the mean.
   - **Histogram Shape**: Bell-shaped, symmetric.
   - **Probability Distribution**: Normal distribution (Gaussian).
   - **Example**: Heights of people in a population.

   ![Normal Distribution](https://upload.wikimedia.org/wikipedia/commons/thumb/8/8c/Standard_deviation_diagram.svg/350px-Standard_deviation_diagram.svg.png)

2. **Exponential Distribution**:
   - **Description**: A continuous distribution with a rapidly decreasing tail.
   - **Histogram Shape**: Right-skewed, one-sided.
   - **Probability Distribution**: Exponential distribution.
   - **Example**: Inter-arrival times in a Poisson process.

   ![Exponential Distribution](https://upload.wikimedia.org/wikipedia/commons/thumb/e/ec/Exponential_pdf.svg/350px-Exponential_pdf.svg.png)

3. **Uniform Distribution**:
   - **Description**: A distribution with a constant probability across an interval.
   - **Histogram Shape**: Rectangular, flat.
   - **Probability Distribution**: Uniform distribution.
   - **Example**: Outcomes of a fair six-sided die.

   ![Uniform Distribution](https://upload.wikimedia.org/wikipedia/commons/thumb/9/96/Uniform_Distribution_PDF_SVG.svg/350px-Uniform_Distribution_PDF_SVG.svg.png)

4. **Binomial Distribution**:
   - **Description**: A discrete distribution for the number of successes in a fixed number of trials.
   - **Histogram Shape**: Discrete, often bimodal.
   - **Probability Distribution**: Binomial distribution.
   - **Example**: Number of heads in a series of coin flips.

   ![Binomial Distribution](https://upload.wikimedia.org/wikipedia/commons/thumb/7/75/Binomial_distribution_pmf.svg/350px-Binomial_distribution_pmf.svg.png)

5. **Poisson Distribution**:
   - **Description**: A discrete distribution modeling the number of events in a fixed interval.
   - **Histogram Shape**: Discrete, often right-skewed.
   - **Probability Distribution**: Poisson distribution.
   - **Example**: Number of customer arrivals at a store in an hour.

   ![Poisson Distribution](https://upload.wikimedia.org/wikipedia/commons/thumb/1/16/Poisson_pmf.svg/350px-Poisson_pmf.svg.png)

6. **Bimodal Distribution**:
   - **Description**: A distribution with two distinct modes or peaks.
   - **Probability Distribution**: A mixture of probability distributions (e.g., two normal distributions).
   
   ![Bimodal Distribution](https://upload.wikimedia.org/wikipedia/commons/e/e2/Bimodal.png)

7. **Skewed Distributions (Left or Right)**:
   - **Description**: A distribution where data is predominantly concentrated on one side.
   - **Probability Distribution**: Skewed distributions, such as the log-normal distribution for right-skewed data.

   ![Skewed Distribution](https://upload.wikimedia.org/wikipedia/commons/thumb/c/cc/Relationship_between_mean_and_median_under_different_skewness.png/868px-Relationship_between_mean_and_median_under_different_skewness.png)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, uniform, binom, poisson

# Data for different distributions
data_normal = np.random.normal(0, 1, 1000)
data_exponential = np.random.exponential(1, 1000)
data_uniform = np.random.uniform(0, 10, 1000)
data_binomial = np.random.binomial(10, 0.5, 1000)
data_poisson = np.random.poisson(3, 1000)
data_bimodal = np.concatenate([np.random.normal(-2, 1, 500), np.random.normal(2, 1, 500)])

# Create subplots for each histogram
plt.figure(figsize=(14, 8))

# Normal distribution
plt.subplot(2, 3, 1)
plt.hist(data_normal, bins=30, density=True, alpha=0.7, color='b', edgecolor='black')
plt.title('Normal Distribution')

# Exponential distribution
plt.subplot(2, 3, 2)
plt.hist(data_exponential, bins=30, density=True, alpha=0.7, color='b', edgecolor='black')
plt.title('Exponential Distribution')

# Uniform distribution
plt.subplot(2, 3, 3)
plt.hist(data_uniform, bins=30, density=True, alpha=0.7, color='b', edgecolor='black')
plt.title('Uniform Distribution')

# Binomial distribution
plt.subplot(2, 3, 4)
plt.hist(data_binomial, bins=np.arange(0, 12) - 0.5, density=True, alpha=0.7, color='b', edgecolor='black')
plt.title('Binomial Distribution')

# Poisson distribution
plt.subplot(2, 3, 5)
plt.hist(data_poisson, bins=30, density=True, alpha=0.7, color='b', edgecolor='black')
plt.title('Poisson Distribution')

# Bimodal distribution
plt.subplot(2, 3, 6)
plt.hist(data_bimodal, bins=30, density=True, alpha=0.7, color='b', edgecolor='black')
plt.title('Bimodal Distribution')

plt.tight_layout()
plt.show()
```

A uniform distribution, also known as a rectangular distribution, is a probability distribution where all outcomes have an equal likelihood of occurring. In a continuous uniform distribution, the probability density is constant over a specified range.

A real-life example of a uniform distribution is the random selection of a number between 1 and 6 when rolling a fair six-sided die. Each number (1, 2, 3, 4, 5, and 6) has an equal probability of 1/6.

```python
# Dice roll modelling
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate random data for simulating dice rolls (1 to 6)
    # random.uniform(low=0.0, high=1.0, size=None) - (includes low, but excludes high).
    # #Draw samples from a uniform distribution.
    # Any value within the given interval is equally likely to be drawn by uniform.

data = np.random.uniform(1, 7, 1000) 

# Create a DataFrame for the data using pandas
df = pd.DataFrame(data, columns=['Dice'])

# Plot a histogram of the generated data
plt.hist(df['Dice'], bins=6, density=True, alpha=0.7, color='b', edgecolor='black')
plt.xlabel('Dice')
plt.ylabel('Probability Density')
plt.title('Uniform Distribution')
plt.show()
```

### Normal or Gaussian distribution

$$ f(x) = \frac{1}{{\sigma \sqrt{2\pi}}} \, e^{-\frac{{(x - \mu)^2}}{{2\sigma^2}}}$$

>f(x): This is the probability density function, which represents the probability of a random variable 'x' having a particular value.
<br>mu(&mu;): The mean (average) of the distribution.
<br>sigma(&sigma;): The standard deviation, which measures the spread or variability of the distribution.
<br>e: The mathematical constant approximately equal to 2.71828.
<br>pi(&pi;): The mathematical constant pi, approximately equal to 3.14159.

A normal (Gaussian) distribution is a probability distribution that forms a symmetric bell-shaped curve. In this distribution, data tends to cluster around the mean, and its shape is characterized by a mean and standard deviation.

Real-life example: Heights of people in a population often follow a normal distribution. Most people are near the average height, and fewer people are much taller or shorter.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate heights of people in a population with a normal distribution
mean_height = 170  # Mean height in centimeters
std_deviation = 10  # Standard deviation

# Generate random data for heights
heights = np.random.normal(mean_height, std_deviation, 1000)

# Create a DataFrame for the data using pandas
df = pd.DataFrame(heights, columns=['Height'])

# Create a figure
plt.figure(figsize=(8, 6))
plt.title('Height Distribution (Histogram and Normal Curve)')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')

# Plot the histogram
plt.hist(df['Height'], bins=20, density=True, alpha=0.7, color='b', edgecolor='black', label='Histogram')

# Demonstrate Normal distribution curve, 
    # Calculate the PDF values for the normal distribution
x = np.linspace(mean_height - 3 * std_deviation, mean_height + 3 * std_deviation, 100)
pdf = norm.pdf(x, mean_height, std_deviation)

# Plot the normal distribution curve
plt.plot(x, pdf, 'r-', label='Normal Curve')

plt.legend()
plt.grid(True)
plt.show()

```

### Binominal Distribution

A binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials (experiments) where each trial has two possible outcomes: success or failure. The distribution is characterized by two parameters:

n: The number of trials, which is a fixed, predetermined value.
p: The probability of success on an individual trial, which remains constant across all trials.
The probability mass function (PMF) of the binomial distribution is given by:

<br>P(X = k) = (n choose k) * p^k * (1 - p)^(n - k)
<br>

>**P(X = k)**: This represents the probability of getting exactly 'k' successes in 'n' trials. <br>
**(n choose k)** : This is the binomial coefficient, often denoted as C(n, k), and it calculates the number of ways to choose 'k' successes from 'n' trials.<br>
**p^k**: This term represents the probability of 'k' successes.<br>
**(1 - p)^(n - k)**: This term represents the probability of 'n - k' failures.<br>

<br>The binomial distribution is widely used to model scenarios involving a fixed number of trials with two possible outcomes and a constant probability of success.

**Example:**
Flipping a coin with 50% probability (p = 0.5), 5 times (n = 5).<br> The binomial distribution can be used to model the probability of getting a specific number of heads (successes) in those 5 coin flips.

<br>
P(X = 3) = (5 choose 3) * (0.5)^3 * (0.5)^(5 - 3)
         = (10) * (0.125) * (0.125)
         = *0.3125*
<br> the probability of getting 3 heads in 5 coin flips is 0.3125




```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate a binomial distribution
n_trials = 10 #10 tries
p_success = 0.5 #50% chance
num_simulations = 10000  # Number of simulations

# Simulate the binomial distribution
results = np.random.binomial(n_trials, p_success, num_simulations)

# Create a histogram
plt.figure(figsize=(8, 6))
# the +2 is added to make sure all possible outcomes are covered
 # np.arange(0, n_trials + 2) creates an array that goes from 0 to 11 
plt.hist(results, bins=np.arange(0, n_trials + 2))

plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xticks(range(n_trials + 1)) # cover all 10 outcomes again
plt.grid(True)
plt.show()
```

The resulting histogram resembles a normal distribution histogram but only because the probability (p) was set to 50%.

>Central Limit Theorem: The Central Limit Theorem states that the sampling distribution of the sample mean from a population with any shape of distribution approaches a normal distribution as the sample size increases. In this case, the number of heads in 10 coin tosses represents a sample mean. As the number of simulations (sample size) increases in the code, the histogram will tend to approximate a normal distribution.

### Geometric and exponential distributions
##### Geometric distribution


The geometric distribution models the number of Bernoulli trials (coin flips) needed to achieve the first success. It is characterized by a single parameter:

p: The probability of success on an individual trial.

**P(X = k) = (1 - p)^(k - 1) * p**

>**P(X = k)**: This represents the probability of achieving the first success on the k-th trial.<br>
**(1 - p)^(k - 1)**: This term represents the probability of 'k - 1' consecutive failures followed by a success.<br>
**p**: This is the probability of success on a single trial.

##### Exponential distribution

The exponential distribution models the time between events in a Poisson process. It is characterized by a single parameter:

λ (lambda): The rate parameter, which represents the average number of events (e.g., arrivals, occurrences) per unit of time.
The probability density function (PDF) of the exponential distribution is given by:<br>
**f(x) = λ * e^(-λx)**

>**f(x)**: This represents the probability density of the random variable 'x.'<br>
**λ**: The rate parameter, which determines the average rate of events.<br>
**e**: The mathematical constant approximately equal to 2.71828.<br>
**x**: The random variable representing the time between events.<br>

Real-Life Example: The exponential distribution can be used to model the time between arrivals of customers at a service center, the time between phone calls to a call center, or the lifetime of electronic components.

In both distributions, the parameters define the shape of the distribution. The geometric distribution is discrete and focuses on the number of trials until the first success, while the exponential distribution is continuous and focuses on the time between events in a Poisson process.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the geometric distribution
p = 0.5  # Probability of success 50%
size = 10000  # Number of trials

# Simulate the geometric distribution
geometric_data = np.random.geometric(p, size)

# Create a histogram for the geometric distribution
plt.figure(figsize=(8, 6))
plt.hist(geometric_data, bins=np.arange(0, geometric_data.max() + 2) - 0.5, density=True, alpha=0.7, color='b', edgecolor='black')
plt.xlabel('Number of Trials until First Success')
plt.ylabel('Probability')
plt.title('Geometric Distribution (p=0.3)')
plt.xticks(range(geometric_data.max() + 1))
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter for the exponential distribution (rate parameter)
lambda_ = 0.2  # Average rate of events

# Simulate the exponential distribution
exponential_data = np.random.exponential(scale=1/lambda_, size=10000)

# Create a histogram for the exponential distribution
plt.figure(figsize=(8, 6))
plt.hist(exponential_data, bins=30, density=True, alpha=0.7, color='b', edgecolor='black')
plt.xlabel('Time between Events')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution (λ=0.2)')
plt.grid(True)
plt.show()
```
